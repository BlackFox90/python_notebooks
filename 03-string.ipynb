# **Strings** *o cadenas de texto*

Las cadenas de texto o *strings* son una lista de caracteres que sirven para almacenar texto. 

```python
saludo = "Hola Miguel!"

# obtener la cantidad de caracteres de la caden

caracteres = len(saludo)

print(caracteres)

# obtener el ultimo caracter de la cadena

print(saludo[-1])
```

## Transformacion de texto

```python
# Convertir la cadena a mayusculas 

mayuscula = saludo.upper()

print(mayuscula)

# Convertir la cadena a minuscula

minuscula = saludo.lower()

print(minuscula)

# Eliminar espacios en blanco en los bordes de la cadena

oracion = " mi perro se comió todas las mandarinas "

oracion_sin_espacios = oracion.strip()

print(oracion_sin_espacios + ".")

# Eliminar caracteres del texto

eliminar_letra = oracion.strip(' m')

print(eliminar_letra + ".")

# Como remplazar caracters en un string
# Encadenamiento de métodos

nueva_oracion = oracion.replace("perro", "gato").replace("mandarinas", "bananas")

print(nueva_oracion)

# Como obtener elementos de una cadena de texto
# Como convertir un texto en una lista de palabras

lista_palabras = oracion_sin_espacios.split(" ")

print(lista_palabras)
```

## Verificación de texto

```python
# Verifica si los caracteres de un STR son alfanúmericos (a-z, A-Z y 0-9).
str.isalnum()

print 'ab123'.isalnum()
True
print 'ab123#'.isalnum()
False

# Verifica si todos son alfabeticos (a-z y A-Z).
str.isalpha()

print 'abcD'.isalpha()
True
print 'abcd1'.isalpha()
False

# Verifica si todos son números (0-9).
str.isdigit()

print '1234'.isdigit()
True
print '123edsd'.isdigit()
False

# Verifica si todos estan en minuscula (a-z).
str.islower()

print 'abcd123#'.islower()
True
print 'Abcd123#'.islower()
False

# Verifica si todos estan en mayuscula (A-Z).
str.isupper()

print 'ABCD123#'.isupper()
True
print 'Abcd123#'.isupper()
False
```

## Tipo de cadena **Multiline**

```python
parrafo = """Yo sé un himno gigante y extraño
    que anuncia en la noche del alma una aurora,
y estas páginas son de este himno
cadencias que el aire dilata en la sombras.
Yo quisiera escribirlo, del hombre
domando el rebelde, mezquino idioma,
con palabras que fuesen a un tiempo
suspiros y risas, colores y notas.
Pero en vano es luchar; que no hay cifra
capaz de encerrarle, y apenas ¡oh hermosa!
si teniendo en mis manos las tuyas
pudiera, al oído, cantártelo a solas."""

print(parrafo)
```

```python
# Obtener si la palabra "colores" se encuentra en el parrafo

existe = "colores" in parrafo

# Cantidad de veces que aparece una palabra en el parrafo

ocurrencias_palabra = parrafo.count("la")

print(existe)
print(ocurrencias_palabra)
```

## Fromateo de cadenas (string format)

```python
edad = 25
nombre = "Raul"
resultado = "Raul tiene 25 años."
concatenacion = nombre + " tiene " + str(edad) + " años."
resultado_format = "{} tiene {} años.".format(nombre, edad)
resultado_interpolacion = f"{nombre} tiene {edad} años."

print(nombre, " tiene ", edad, " años.")
print(concatenacion)
print(resultado_format)
print(resultado_interpolacion)
```
